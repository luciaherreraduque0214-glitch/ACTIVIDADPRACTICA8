Crear un análisis simple de las ventas de un pequeño negocio, incluyendo la
creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados
con Matplotlib.  
Los estudiantes trabajarán con los siguientes datos de ventas mensuales:  
Datos de ventas mensuales (en miles)  
data = {“Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto",
"Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60,
70, 80, 75, 90],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas mensuales proporcionados
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

# 1. Creación del DataFrame
df = pd.DataFrame(data)
print("DataFrame de Ventas:")
display(df.head())

# 2. Operaciones de NumPy para análisis de ventas
total_ventas = np.sum(df['Ventas'])
promedio_ventas = np.mean(df['Ventas'])
min_ventas = np.min(df['Ventas'])
max_ventas = np.max(df['Ventas'])

print(f"\nAnálisis de Ventas (en miles):")
print(f"  Total de Ventas Anuales: {total_ventas}")
print(f"  Promedio de Ventas Mensuales: {promedio_ventas:.2f}")
print(f"  Ventas Mínimas (Mes con menos ventas): {min_ventas}")
print(f"  Ventas Máximas (Mes con más ventas): {max_ventas}")

# 3. Visualización de los resultados con Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(df['Mes'], df['Ventas'], color='skyblue')
plt.xlabel('Mes')
plt.ylabel('Ventas (en miles)')
plt.title('Ventas Mensuales del Negocio')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Ajustar el diseño para evitar que las etiquetas se corten
plt.show()
```

 Realizar un análisis simple de las calificaciones de estudiantes en diferentes
asignaturas, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la
visualización de los resultados con Matplotlib.
Los estudiantes trabajarán con los siguientes datos de calificaciones:  
Datos de calificaciones de estudiantes  
data = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah",
"Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80,
88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes proporcionados
data = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# 1. Creación del DataFrame
df_calificaciones = pd.DataFrame(data)
print("DataFrame de Calificaciones de Estudiantes:")
display(df_calificaciones.head())
```

```python
# 2. Operaciones de NumPy para análisis de calificaciones
print("\nAnálisis de Calificaciones por Asignatura:")
for materia in ['Matemáticas', 'Ciencias', 'Historia']:
    promedio_materia = np.mean(df_calificaciones[materia])
    max_materia = np.max(df_calificaciones[materia])
    min_materia = np.min(df_calificaciones[materia])
    print(f"  {materia}:")
    print(f"    Promedio: {promedio_materia:.2f}")
    print(f"    Calificación Máxima: {max_materia}")
    print(f"    Calificación Mínima: {min_materia}")

# Promedio general por estudiante
df_calificaciones['Promedio General'] = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean(axis=1)
print("\nPromedio General por Estudiante:")
display(df_calificaciones[['Estudiante', 'Promedio General']].head())
```

```python
# 3. Visualización de los resultados con Matplotlib

# Gráfico de barras de promedios por asignatura
promedio_asignaturas = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean()

plt.figure(figsize=(10, 6))
plt.bar(promedio_asignaturas.index, promedio_asignaturas.values, color=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.xlabel('Asignatura')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Asignatura')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Gráfico de barras de promedios generales por estudiante
plt.figure(figsize=(12, 6))
plt.bar(df_calificaciones['Estudiante'], df_calificaciones['Promedio General'], color='salmon')
plt.xlabel('Estudiante')
plt.ylabel('Promedio General')
plt.title('Promedio General de Calificaciones por Estudiante')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
