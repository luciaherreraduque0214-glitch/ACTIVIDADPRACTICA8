Actividad 1: Juego de adivinar el número
Pasos:
1. Importar el módulo random para generar números aleatorios.
2. Crear una función generar_numero() que retorne un número aleatorio entre 1 y 100.
3. Implementar un bucle while para el juego principal.
4. Dentro del bucle, usar try-except para manejar entradas no numéricas.
5. Usar if-elif-else para dar pistas al jugador (mayor, menor o correcto).
6. Implementar un contador de intentos.
7. Usar break para salir del bucle cuando se adivine el número.

```python
import random

def generar_numero():
    """Genera un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)

def jugar_adivina_numero():
    numero_secreto = generar_numero()
    intentos = 0
    print("¡Bienvenido al juego de adivinar el número!")
    print("He pensado en un número entre 1 y 100. ¡Intenta adivinarlo!")

    while True:
        try:
            intento_usuario = int(input("Introduce tu intento: "))
            intentos += 1

            if intento_usuario < numero_secreto:
                print("¡Demasiado bajo! Intenta con un número mayor.")
            elif intento_usuario > numero_secreto:
                print("¡Demasiado alto! Intenta con un número menor.")
            else:
                print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
                break
        except ValueError:
            print("Entrada inválida. Por favor, introduce un número entero.")

# Para iniciar el juego, puedes llamar a la función:
jugar_adivina_numero()
```

Actividad 2: Calculadora de operaciones básicas
Pasos:
1. Definir funciones para cada operación (suma, resta, multiplicación, división).
2. Crear un menú de opciones usando un bucle while.
3. Utilizar if-elif-else para seleccionar la operación.
4. Implementar try-except para manejar errores (como división por cero).
5. Usar continue para volver al menú principal después de cada operación.
6. Implementar una opción para salir del programa usando break.

```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ZeroDivisionError("No se puede dividir por cero.")
    return a / b

def calculadora():
    while True:
        print("\n--- Calculadora de Operaciones Básicas ---")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Salir")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == '5':
            print("¡Hasta luego!")
            break

        if opcion in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Introduce el primer número: "))
                num2 = float(input("Introduce el segundo número: "))
            except ValueError:
                print("Entrada inválida. Por favor, introduce números válidos.")
                continue

            try:
                if opcion == '1':
                    print(f"Resultado: {num1} + {num2} = {suma(num1, num2)}")
                elif opcion == '2':
                    print(f"Resultado: {num1} - {num2} = {resta(num1, num2)}")
                elif opcion == '3':
                    print(f"Resultado: {num1} * {num2} = {multiplicacion(num1, num2)}")
                elif opcion == '4':
                    print(f"Resultado: {num1} / {num2} = {division(num1, num2)}")
            except ZeroDivisionError as e:
                print(f"Error: {e}")
            except Exception as e:
                print(f"Ocurrió un error inesperado: {e}")
        else:
            print("Opción no válida. Por favor, elige una opción del 1 al 5.")

        # Usa continue implícitamente al final del bucle para volver al menú

# Para iniciar la calculadora, puedes llamar a la función:
calculadora()
```

Actividad 3: Lista de tareas
Pasos:  
1. Crear una lista vacía para almacenar las tareas.
2. Definir funciones para agregar, eliminar, y mostrar tareas.
3. Implementar un menú principal usando un bucle while.
4. Usar if-elif-else para manejar las opciones del menú.
5. Utilizar un bucle for para mostrar todas las tareas.
6. Implementar try-except para manejar errores al eliminar tareas.
7. Usar break para salir del programa.

```python
tareas = []

def agregar_tarea(tarea):
    tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea(indice):
    try:
        tarea_eliminada = tareas.pop(indice - 1) # Restar 1 porque los usuarios piensan en índices base 1
        print(f"Tarea '{tarea_eliminada}' eliminada.")
    except IndexError:
        print("Error: Índice de tarea no válido.")
    except ValueError:
        print("Entrada inválida. Por favor, introduce un número entero.")

def mostrar_tareas():
    if not tareas:
        print("No hay tareas en la lista.")
        return
    print("\n--- Tu Lista de Tareas ---")
    for i, tarea in enumerate(tareas):
        print(f"{i + 1}. {tarea}")
    print("-------------------------")

def lista_de_tareas():
    while True:
        print("\n--- Menú Lista de Tareas ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Mostrar tareas")
        print("4. Salir")

        opcion = input("Selecciona una opción (1-4): ")

        if opcion == '1':
            tarea = input("Introduce la nueva tarea: ")
            agregar_tarea(tarea)
        elif opcion == '2':
            mostrar_tareas()
            if tareas:
                try:
                    indice_str = input("Introduce el número de la tarea a eliminar: ")
                    indice = int(indice_str)
                    eliminar_tarea(indice)
                except ValueError:
                    print("Entrada inválida. Por favor, introduce un número entero.")
        elif opcion == '3':
            mostrar_tareas()
        elif opcion == '4':
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, elige una opción del 1 al 4.")

# Para iniciar la aplicación de la lista de tareas, puedes llamar a la función:
lista_de_tareas()


```
